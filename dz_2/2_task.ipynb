#### Задание 2. 
Составьте шаблон дизайна эксперимента для гипотезы, которая набрала больше всего баллов в практическом задании предыдущего урока.

Гипотеза:
При добавлении возмозможности кастомизации приложения по собственному вкусу, при оформленной подписке, количество пользователей оформивших подписку увеличится.

Что делаем:
Контрольная версия - На момент запуска эксперимента делаем помесячный подсчёт количества пользователей оформивших подписку за год.
Тестовая версия - Пользователям оформляющим подписку с момента начала эксперимента и пользователям с уже оформленной подпиской добавляем функционал о кастомизации элементов окна месенджера (цветовых схем и \ или установки фото, рисунков как фонов), шрифтов сообщений. Данных пользователей оповещаем сообщением о добавленном функционале и прикладываем ссылку на короткое видео о том каким образом можно настроить тот или иной элемент окна месенджера. Видео должно быть размещено на открытом ресурсе.

На каких пользователях тестируем:
Тестируем на всех пользователях которые оформляют или уже оформили подписку.

Метрики:
Конверсия пользователей оформивших подписку за месяц.
Конверсия пользователей продливших подписку за месяц.
Сравнительный анализ ежемесячный конверсий пользователей оформивших подписку с коверсией пользователей оформивших подписку за аналогичный месяц в предыдущем году.

План действий:
Если наш эксперимент будет положительным и мы зафиксируем ожидаемое улучшение в ключевых метриках, то продолжим поддерживать и улучшать данный функционал и устранять неисправности данного функционала.

Если основные метрики падают или растут недостаточно, либо добавочные метрики падают, то данный функционал оставляем без поддержки.
